```python
import pandas as pd
```

```python
df = pd.read_csv('df_absention_cleaned_14_06_2024__11_01.csv')
```

```python
df.head()
```

```python
# test = df[['year','Greece']]
# test
```

```python
df1 = df.diff(periods = 1)
df1 = df1.drop('year', axis=1)

df1
```

```python
df_year = pd.DataFrame(df['year'])
df_year
```

```python
df_final = pd.concat([df_year, df1], axis = 1)
df_final
```

```python
df_final = df_final.T.reset_index()
df_final
```

```python
df_final.columns = df_final.iloc[0]
df_final = df_final[1:]
df_final
```

```python
df_final = df_final.sort_values(by=2024.0,ascending=False)
df_final
```

```python
top = df_final.head(11)
top = top[['year',2024.0]]
top = top.rename(columns = {'year':'country',
                           2024.0:'change'})
top
```

```python
top.to_csv('change.csv', index=False)
```
