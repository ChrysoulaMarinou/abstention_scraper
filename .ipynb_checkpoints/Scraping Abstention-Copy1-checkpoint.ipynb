```python
#imports
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np

#set ups
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
url = 'https://results.elections.europa.eu/en/turnout/'
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
#finding the table i wnat to scrape
table = soup.find_all('table',{"class":"turnout-by-country--table basic-table"})
```

```python
year_col = []

for t in table:
    heads = t.find_all('thead')  # Find all 'thead' elements
    for head in heads:
        columns = head.find_all('th')  # Find all 'th' elements within the 'thead'
        # Extract text from each 'th' element and append to 'headers' list
        for column in columns:
            year_col.append(column.text.strip())  # Strip any extra whitespace
```

```python
mapping = dict()

for t in table:
    main_body = t.find('tbody')
    
    if main_body:
        rows = main_body.find_all('tr')
        country = main_body.find_all('span', {'class':"turnout-by-country--country-name--INTERACTIVE"})
        countries = [span.text.strip() for span in country]    
                  
        index = 0
        for row in rows:
            cells = row.find_all('td')
            turnout = []
            
            for cell in cells:
                text = cell.text.strip()
                
                if not text:
                    text = 'NaN'
                turnout.append(text)
                
            mapping[countries[index]] = turnout
            
            index += 1
        print(mapping) 
```

```python
df = pd.DataFrame(mapping)
df['year'] = year_col[1:]
df = df[['year','Belgium', 'Denmark', 'Germany', 'Ireland', 'France', 'Italy',
       'Luxembourg', 'Netherlands', 'United Kingdom', 'Greece', 'Spain',
       'Portugal', 'Sweden', 'Austria', 'Finland', 'Czechia', 'Estonia',
       'Cyprus', 'Lithuania', 'Latvia', 'Hungary', 'Malta', 'Poland',
       'Slovenia', 'Slovakia', 'Bulgaria', 'Romania', 'Croatia', 'Total EU']]
```

```python
df
```

```python
df_abstention = df.astype(float).copy()
df_abstention.iloc[:, 1:] = df_abstention.iloc[:, 1:].applymap(lambda x: 100 - x).copy()
df_abstention ['year'] = df_abstention ['year'].astype(int).copy()
df_abstention 
```

```python
article = soup.find('div',{'class':"last-update"})
info = article.find('strong').text.strip()
info = info.replace('/', '_').replace(' ', '_').replace('-','')
info
```

```python
df_abstention.to_csv(f'abs_turnout_{info}.csv',index=False)
```

```python
df.to_csv(f'turnout_{info}.csv',index=False)
```
